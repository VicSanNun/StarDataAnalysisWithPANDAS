```python
import pandas as pd

stars = pd.read_csv("StarPredict_.csv")

pd.set_option("display.max_rows", 10)
```

### 3 - Select the first row of data (the first record) from stars, assigning it to the variable first_row.

```python
first_row = stars.iloc[0]

pd.DataFrame(first_row)
```

### 5 - Select the records with index labels 1, 2, 3, 5, and 8, assigning the result to the variable sample_reviews.

```python
sample_reviews = stars.iloc[[1, 2, 3, 5, 8]]

pd.DataFrame(sample_reviews)
```

### 6 - Create a variable df containing the Temperature (K)	, Absolute magnitude(Mv), Luminosity(L/Lo), and Spectral Class columns of the records with the index labels 0, 1, 10, and 100.

```python
df = stars.loc[[0, 1, 10, 100], ['Temperature', 'Absolute magnitude', 'Luminosity', 'Spectral Class']]

pd.DataFrame(df)
```

### 7 - Create a variable df containing the Radius(R/Ro) and Star color columns of the first 100 records.

```python
df = stars.loc[0:99, ['Radius', 'Star color']]

pd.DataFrame(df)
```

### 8 - Create a DataFrame smaller_sun_radius containing stars of Radius < 1:

```python
smaller_sun_radius = stars.loc[stars.Radius < 1]

pd.DataFrame(smaller_sun_radius)
```
